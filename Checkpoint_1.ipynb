```python
from google.colab import drive
import zipfile
import pandas as pd

# 1. Mount Google Drive
drive.mount('/content/drive')

# 2. Unzip the dataset
zip_path = '/content/drive/My Drive/Colab Notebooks/CSCE 676/Project/online+retail+ii.zip'
extract_path = '/content/drive/My Drive/Colab Notebooks/CSCE 676/Project/online_retail_ii'

with zipfile.ZipFile(zip_path, 'r') as z:
    z.extractall(extract_path)

# 3. Load the UCI Excel file
file_path = extract_path + '/online_retail_II.xlsx'
df = pd.read_excel(file_path)

df.head()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Ensure correct types
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"], errors="coerce")
df = df.dropna(subset=["Invoice", "StockCode"])
df = df[df["Quantity"] > 0]

# --------------------------------------------------
# 1. Distribution of basket sizes
# --------------------------------------------------
basket_sizes = df.groupby("Invoice")["StockCode"].nunique()

plt.figure(figsize=(6,4))
plt.hist(basket_sizes, bins=30, edgecolor="black")
plt.title("Basket Size Distribution")
plt.xlabel("Unique items per invoice")
plt.ylabel("Frequency")
plt.show()

# --------------------------------------------------
# 2. Frequency of top items
# --------------------------------------------------
item_freq = df["StockCode"].value_counts().head(20)

plt.figure(figsize=(6,4))
item_freq.plot(kind="bar")
plt.title("Top 20 Most Frequent Items")
plt.xlabel("Item")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# --------------------------------------------------
# 3. Sparsity of item co-occurrence
# --------------------------------------------------
# Use top N items to keep matrix manageable
top_items = df["StockCode"].value_counts().head(50).index
df_top = df[df["StockCode"].isin(top_items)]

# Invoice-item matrix
invoice_item = pd.crosstab(df_top["Invoice"], df_top["StockCode"])

# Co-occurrence matrix
co_matrix = invoice_item.T.dot(invoice_item)

# Sparsity calculation
total_cells = co_matrix.size
nonzero_cells = np.count_nonzero(co_matrix.values)
sparsity = 1 - (nonzero_cells / total_cells)

print("Co-occurrence matrix sparsity:", round(sparsity, 4))

plt.figure(figsize=(6,5))
plt.imshow(co_matrix, cmap="Blues")
plt.title("Co-occurrence Heatmap (Top 50 Items)")
plt.colorbar()
plt.tight_layout()
plt.show()

# --------------------------------------------------
# 4. Temporal gaps between transactions
# --------------------------------------------------
df_sorted = df.sort_values("InvoiceDate")
df_sorted["time_gap_hours"] = df_sorted["InvoiceDate"].diff().dt.total_seconds() / 3600.0

plt.figure(figsize=(6,4))
plt.hist(df_sorted["time_gap_hours"].dropna(), bins=50, edgecolor="black")
plt.title("Distribution of Time Gaps Between Transactions (Hours)")
plt.xlabel("Hours")
plt.ylabel("Frequency")
plt.xlim(0, 48)  # focus on short-term gaps
plt.show()
```

**Distribution of Basket Sizes**

The distribution of unique items per invoice is highly right-skewed.  
Most baskets contain only a small number of items, with the majority under 10 unique items.  
A long tail of larger baskets exists but is rare.  
This skew suggests that support thresholds must be chosen carefully to avoid eliminating meaningful but infrequent patterns.

**Frequency of Top Items**

The top 20 most frequent items show a strong popularity imbalance.  
A few items (e.g., 85123A) appear thousands of times, while most items appear far less often.  
This long-tail distribution indicates that naive association rule mining will be dominated by a small set of high-frequency items.

**Sparsity of Item Co-occurrence**

The co-occurrence matrix for the top 50 items is extremely sparse.  
Most item pairs never appear together, and the matrix is dominated by diagonal values (self-co-occurrence).  
This sparsity motivates external techniques such as:
- dimensionality reduction,
- item embeddings,
- or sequential pattern mining, which can capture structure beyond simple co-occurrence.

**Temporal Gaps Between Transactions**

Time gaps between transactions are concentrated in short intervals, mostly under 10 hours.  
This indicates bursts of purchasing activity followed by longer periods of inactivity.  
The presence of temporal structure supports the use of external techniques such as sequential pattern mining or time-aware models.

**Initial Observations Motivating External Techniques**

- Skewed basket sizes and item frequencies suggest that standard support-based itemset mining may miss important patterns.
- Extreme sparsity in co-occurrence indicates that richer representations (embeddings, sequences) may capture more meaningful relationships.
- Temporal patterns justify exploring sequential pattern mining beyond the course material.
