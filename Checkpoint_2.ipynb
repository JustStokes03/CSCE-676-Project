# 1) Project Scope

This project analyzes a retail transaction dataset to uncover meaningful item relationships, temporal patterns, and structural properties of customer purchasing behavior. Core course techniques include frequent itemset mining and association rule analysis. EDA revealed highly skewed basket sizes, long-tail item frequencies, and extreme sparsity in item co-occurrence, motivating the use of external methods such as sequential pattern mining, dimensionality reduction, and item embeddings. Temporal gap analysis further supports exploring time-aware or sequence-based models to capture ordering effects that standard frequent itemsets may miss.

# 2) Research Question Definition

## RQ1 (Course Technique)
**Research Question:**  
What frequent itemsets emerge under varying support thresholds, and how does threshold choice influence the diversity and interpretability of discovered patterns?

**Data Mining Task Type:**  
Frequent itemset mining

**Relevant Algorithms:**  
Apriori, FP-Growth

**Evaluation Criteria:**  
Support, confidence, lift, number of itemsets discovered, interpretability


## RQ2 (Course Technique)
**Research Question:**  
How do confidence and lift differ in evaluating the strength and usefulness of association rules derived from the dataset's highly skewed item frequency distribution?

**Data Mining Task Type:**  
Association rule mining

**Relevant Algorithms:**  
Apriori-based rule generation, FP-Growth rule generation

**Evaluation Criteria:**  
Confidence, lift, conviction, rule redundancy, coverage


## RQ3 (External Technique)
**Research Question:**  
Do sequential patterns reveal temporal or ordering structure that is missed by unordered frequent itemsets, particularly given the bursty transaction timing and sparse co-occurrence matrix?

**Data Mining Task Type:**  
Sequential pattern mining

**Relevant Algorithms:**  
PrefixSpan, SPADE, time-aware sequence models (optional)

**Evaluation Criteria:**  
Pattern frequency, temporal coherence, interpretability, comparison with unordered itemsets

# 3) Motivation and Feasibility

**Motivation**  
EDA reveals several structural challenges in the dataset:  
- Basket sizes are highly right-skewed, meaning most transactions contain very few items.  
- Item frequencies follow a long-tail distribution, with a handful of dominant items.  
- The co-occurrence matrix is extremely sparse, limiting the effectiveness of simple support-based mining.  
- Temporal gaps cluster within short intervals, indicating bursty purchasing behavior.  

These patterns motivate research questions that go beyond standard frequent itemset mining and incorporate ordering, temporal structure, and richer representations.

**Non-Triviality**  
- Course techniques (Apriori, FP-Growth) ignore ordering and temporal structure.  
- Sequential pattern mining and embedding-based methods can capture structure that unordered itemsets cannot.  
- Varying support thresholds introduces trade-offs between pattern diversity and noise.

**Feasibility**  
- Apriori and FP-Growth are directly available in standard Python libraries (mlxtend, efficient implementations).  
- PrefixSpan and SPADE have Python implementations with manageable computational cost for this dataset size.  
- Preliminary EDA confirms that sequence lengths and temporal density are sufficient for sequential mining.  
- Dimensionality reduction and embedding methods (e.g., item2vec) are well-supported and scalable.

**Risks**  
- Sequential mining may produce large pattern sets if parameters are not tuned.  
- Embedding models require careful hyperparameter selection and may overfit sparse data.  
- Support thresholds that are too low may lead to combinatorial explosion.


# 4) Methodological Planning

**Course Algorithms**  
- **Apriori**: Baseline frequent itemset mining; useful for understanding support-threshold sensitivity.  
- **FP-Growth**: More scalable alternative; used to validate Apriori results and reduce runtime.  
- **Association Rule Generation**: Used to compare confidence, lift, conviction, and rule redundancy.

**External Algorithms**  
- **PrefixSpan**: Primary sequential pattern mining algorithm to capture ordering effects.  
- **SPADE**: Alternative sequence miner for cross-validation of sequential structure.  
- **Optional**: Item embeddings (item2vec) or dimensionality reduction to address co-occurrence sparsity.

**Evaluation Plan**  
- **Frequent Itemsets**: Support, confidence, lift, number/diversity of itemsets.  
- **Association Rules**: Confidence vs. lift comparison, redundancy, coverage.  
- **Sequential Patterns**: Temporal coherence, pattern frequency, interpretability, comparison with unordered itemsets.  
- **Embeddings (if used)**: Cosine similarity structure, cluster coherence.

**Baselines**  
- High-support-only frequent itemset mining (course-only baseline).  
- Unordered itemsets vs. sequential patterns.  
- Raw co-occurrence vs. embedding-based similarity.

**Initial Feasibility Checks**  
- Verified that Apriori/FP-Growth run efficiently on subsampled data.  
- Confirmed that transaction sequences are long enough for PrefixSpan to produce non-trivial patterns.  
- Checked that timestamp granularity supports temporal ordering.
